### Read disease.csv to data (dataframe)

```python
data = pd.read_csv("disease.csv")
```

### Cleaning Process

```python
data.columns = ["extra","Disease","Count","Symptoms"]        # Change Columns names in DF
```

```python
data = data[["Disease","Symptoms"]]                  # select only needed columns in Dataframe
data.head(5)                                                         # uncleaned dataframe
```

```python
data1 = data
data1["Disease"]=data["Disease"].fillna(method='ffill')         # Fill the NaN's with the above value
```

```python
# Remove first row

data1 = data1.iloc[1:]                               # Remove the first row ( dublicate )
data1
```

```python
kk=data1
```

```python
kk.Disease = kk.Disease.str.replace("^","_")
kk.Symptoms=kk.Symptoms.str.replace("^","_")
```

```python
kk = kk.apply(lambda x: x.str.replace("_",","))
```

```python
kk.head(10)
```

```python
a = kk["Disease"].str.split(",", n = 1, expand = True)
```

```python
a = a[1].str.split(",", n = 1, expand = True)
a[0] = a[0].replace('  ',' ')
```

```python
b = kk["Symptoms"].str.split(",", n = 1, expand = True)
b = b[1].str.split(",", n = 1, expand = True)
b[0] = b[0].replace('  ',' ')
```

```python
data['Disease']=a[0].astype(str)
data['Symptoms']=b[0].astype(str)
data['Disease']=data['Disease'].replace('\s+', ' ', regex=True)
data['Symptoms']=data['Symptoms'].replace('\s+', ' ', regex=True)
data['Disease']=data['Disease'].replace('-', ' ')
data['Symptoms']=data['Symptoms'].replace('-', ' ')
```

```python
data = data.iloc[1:]
```

### Final dataset cleaned

```python
data.to_csv("final_disease.csv", index=False)

data.head(15)    # 15 rows to show
```
